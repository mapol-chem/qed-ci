```python
import numpy as np
import psi4
psi4.core.set_output_file("HeHp_Monomer_FCI_ccpVDZ.out")
```

```python
mol_str = """
2 1
H 0.0000000000   0.0000000000  -5000000.38
He 0.0000000000   0.0000000000  -4999999.62
H 0.0000000000   0.0000000000  4999999.62
He 0.0000000000   0.0000000000  5000000.38
symmetry c1
"""
#-100.38
#-99.62
#99.62
#100.38
#mol_str = """
#1 1
#H 0.0000000000   0.0000000000  11999.62869866
#He 0.0000000000   0.0000000000  12000.37130134
#H 0.0000000000   0.0000000000 -12000.37130134
#He 0.0000000000   0.0000000000 -11999.62869866
#symmetry c1
#"""

options_dict = {
    "basis": "cc-pVDZ",
    "scf_type": "pk",
    "e_convergence": 1e-10,
    "d_convergence": 1e-10,
    'num_roots' : 10,
    'frozen_docc' : [0],
    'TDM' :  True,
    'OPDM': True,
    'dipmom' : True
}

mol = psi4.geometry(mol_str)
psi4.set_options(options_dict)
scf_e, wfn = psi4.energy('SCF', return_wfn=True)
fci_energy, wfn = psi4.energy('fci',ref_wfn=wfn, return_wfn=True)
print(fci_energy)
#props = ['DIPOLE', 'TRANSITION_DIPOLE']
#psi4.properties('fci',properties=props)
Edimer = 2 * -2.9603712198036938
print(fci_energy - Edimer)
print(fci_energy - -5.92074238660101)
```

```python

#transition density matrix
opdm_0 = wfn.get_opdm(1,4,"SUM", True)
dm = opdm_0.to_array()
print(dm)


#two electron integral
mints = psi4.core.MintsHelper(wfn.basisset())
I = mints.ao_eri()
D = I.to_array()
print(I.np)

#get transition dipole moment
props = ['DIPOLE', 'TRANSITION_DIPOLE']
psi4.properties('fci',properties=props)
M = ['X','Y','Z']
for i in range(6):
        for j in M:
                dipole=psi4.get_variable('CI ROOT %d -> ROOT %d DIPOLE %s'%(i,i+1,j))
                print('CI ROOT %d -> ROOT %d DIPOLE %s = %s'%(i,i+1,j,dipole))
```
